```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./datasets/clustering_data.csv')
df.head
```

```python
df.query('StateName == "MADHYA PRADESH"', inplace = True)
df.shape
```

```python
#8284 rows dropeed this way
for index, row in df.iterrows():
    try:
        str = row['Latitude']
        if(str[-1] == '-'):
            df = df.drop(index=index, axis=0)
            print('dropped')
    except:
        print('exception occured')

for index, row in df.iterrows():
    try:
        str = row['Longitude']
        if(str[-1] == '-'):
            df = df.drop(index=index, axis=0)
            print('dropped')
    except:
        print('exception occured')


df.shape
```

```python

df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] = df['Longitude'].astype(float)
```

```python
for index, row in df.iterrows():
    if(row['Latitude']>26.31 or row['Latitude']<21.5):
        df = df.drop(index=index, axis=0)
        print('dropped')

for index, row in df.iterrows():
    if(row['Longitude']>82.5 or row['Longitude']<74.8):
        df = df.drop(index=index, axis=0)
        print('dropped')

df.shape
        
```

```python
import cv2
ax = plt.subplot()
ax.hexbin(df['Latitude'],df['Longitude'], gridsize=10, cmap='Oranges')
plt.axis('off')
plt.show()
```

```python
import kmeans
dfplot = df.loc[:,['Latitude','Longitude']]
latlong = np.array(dfplot)
```

```python
kmeansinst = kmeans.kmeansclustering(k=20)
labels = kmeansinst.fit(latlong)
plt.scatter(latlong[:,0],latlong[:,1], c=labels, s=5)
plt.scatter(kmeansinst.centroids[:, 0], kmeansinst.centroids[:,1], c=range(len(kmeansinst.centroids)), marker='*')

```
